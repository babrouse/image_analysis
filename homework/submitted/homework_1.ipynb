### Import packages and open the images
We also do some figure display stuff so we can look at them side by side

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_A_png = plt.imread('images/AuLait_gray.png')
image_A_tif = plt.imread('images/AuLait_gray.tif')
```

```python
# Create empty figure with 1 row and two columns then display the images individually
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].imshow(image_A_png, 'gray')
axes[0].set_title('PNG')

axes[1].set_title('TIF')
axes[1].imshow(image_A_tif, 'gray')

plt.tight_layout()  # Ensures they don't overlap

plt.savefig('images/side_by_side.png')
plt.show()
```

At this point they look thee same so we're instructed to print the minimum and maximum values of each array.

```python
print(np.max(image_A_png), np.max(image_A_tif)) 
print(np.min(image_A_png), np.min(image_A_tif))
print(type(image_A_png[1][1]), type(image_A_tif[1][1]))
```

 Next we have to save them as TIFF files using two different methods:

```python
plt.imsave('images/AuLait_png_out.tif', image_A_png, format='TIFF')
```

```python
from skimage import io
io.imsave('images/AuLait_png_out_sk.tif', image_A_png)
```

Now I need to reload these new TIF files back into python to check their shapes.

```python
img_B_tif = io.imread('images/AuLait_png_out.tif')
img_B_tif_sk = io.imread('images/AuLait_png_out_sk.tif')

print(img_B_tif.shape)
print(img_B_tif_sk.shape)
```
